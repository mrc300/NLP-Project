```python
##IMPORTS
import pandas as pd
import re
```

```python
#Week 36

splits = {'train': 'train.parquet', 'validation': 'validation.parquet'}
df_train = pd.read_parquet("hf://datasets/coastalcph/tydi_xor_rc/" + splits["train"])
df_val = pd.read_parquet("hf://datasets/coastalcph/tydi_xor_rc/" + splits["validation"])

arabic_data_size_train = df_train[df_train["lang"] == "ar"].shape[0]
korean_data_size_train = df_train[df_train["lang"] == "ko"].shape[0]
telugu_data_size_train = df_train[df_train["lang"] == "te"].shape[0]

arabic_data_size_val = df_val[df_val["lang"] == "ar"].shape[0]
korean_data_size_val = df_val[df_val["lang"] == "ko"].shape[0]
telugu_data_size_val = df_val[df_val["lang"] == "te"].shape[0]

print('Arabic data size (train): ', arabic_data_size_train)
print('Korean data size (train): ', korean_data_size_train)
print('Telugu data size (train): ', telugu_data_size_train)

print('Arabic data size (val): ', arabic_data_size_val)
print('Korean data size (val): ', korean_data_size_val)
print('Telugu data size (val): ', telugu_data_size_val)


arabic_word_count_train = df_train[df_train["lang"] == "ar"]["question"].apply(lambda x: len(x.split())).sum()
korean_word_count_train = df_train[df_train["lang"] == "ko"]["question"].apply(lambda x: len(x.split())).sum()
telugu_word_count_train = df_train[df_train["lang"] == "te"]["question"].apply(lambda x: len(x.split())).sum()

arabic_word_count_val = df_val[df_val["lang"] == "ar"]["question"].apply(lambda x: len(x.split())).sum()
korean_word_count_val = df_val[df_val["lang"] == "ko"]["question"].apply(lambda x: len(x.split())).sum()
telugu_word_count_val = df_val[df_val["lang"] == "te"]["question"].apply(lambda x: len(x.split())).sum()

print('Word count in Arabic questions: ', arabic_word_count_train)
print('Word count in Korean questions: ', korean_word_count_train)
print('Word count in Telugu questions: ', telugu_word_count_train)

print('Word count in arabic questions (val):', arabic_word_count_val)
print('Word count in korean questions (val):', korean_word_count_val)
print('Word count in telugu questions (val):', telugu_word_count_val)


arabic_avg_word_count_train = arabic_word_count_train / arabic_data_size_train
korean_avg_word_count_train = korean_word_count_train / korean_data_size_train
telugu_avg_word_count_train = telugu_word_count_train / telugu_data_size_train

arabic_avg_word_count_val = arabic_word_count_val / arabic_data_size_train
korean_avg_word_count_val = korean_word_count_val / korean_data_size_train
telugu_avg_word_count_val = telugu_word_count_val / telugu_data_size_train

print('Average word count in Arabic questions (train): ', arabic_avg_word_count_train)
print('Average word count in Korean questions (train): ', korean_avg_word_count_train)
print('Average word count in Telugu questions (Train): ', telugu_avg_word_count_train)

print('Average word count in Arabic questions (val): ', arabic_avg_word_count_val)
print('Average word count in Korean questions (val): ', korean_avg_word_count_val)
print('Average word count in Telugu questions (val): ', telugu_avg_word_count_val)


total_number_arabic_answerable_train = df_train[(df_train["lang"] == "ar") & df_train["answerable"]].shape[0]
total_number_korean_answerable_train = df_train[(df_train["lang"] == "ko") & df_train["answerable"]].shape[0]
total_number_telugu_answerable_train = df_train[(df_train["lang"] == "te") & df_train["answerable"]].shape[0]

total_number_arabic_answerable_val = df_val[(df_val["lang"] == "ar") & df_val["answerable"]].shape[0]
total_number_korean_answerable_val = df_val[(df_val["lang"] == "ko") & df_val["answerable"]].shape[0]
total_number_telugu_answerable_val = df_val[(df_val["lang"] == "te") & df_val["answerable"]].shape[0]

print('Total number of answerable questions in Arabic (train): ', total_number_arabic_answerable_train)
print('Total number of answerable questions in Korean (train): ', total_number_korean_answerable_train)
print('Total number of answerable questions in Telugu (train): ', total_number_telugu_answerable_train)

print('Total number of answerable questions in Arabic (val): ', total_number_arabic_answerable_val)
print('Total number of answerable questions in Korean (val): ', total_number_korean_answerable_val)
print('Total number of answerable questions in Telugu (val): ', total_number_telugu_answerable_val)


ratio_arabic_answerable_train = total_number_arabic_answerable_train / arabic_data_size_train
ratio_korean_answerable_train = total_number_korean_answerable_train / korean_data_size_train
ratio_telugu_answerable_train = total_number_telugu_answerable_train / telugu_data_size_train

ratio_arabic_answerable_val = total_number_arabic_answerable_val / arabic_data_size_train
ratio_korean_answerable_val = total_number_korean_answerable_val / korean_data_size_train
ratio_telugu_answerable_val = total_number_telugu_answerable_val / telugu_data_size_train

print('Ratio of answerable questions in Arabic (train): ', ratio_arabic_answerable_train)
print('Ratio of answerable questions in Korean (train): ', ratio_korean_answerable_train)
print('Ratio of answerable questions in Telugu (train): ', ratio_telugu_answerable_train)

print('Ratio of answerable questions in Arabic (val): ', ratio_arabic_answerable_val)
print('Ratio of answerable questions in Korean (val): ', ratio_korean_answerable_val)
print('Ratio of answerable questions in Telugu (val): ', ratio_telugu_answerable_val)


first_100_arabic_questions_train = df_train[df_train["lang"] == "ar"]["question"].head(100).tolist()
print('First 100 Arabic questions (train): ', first_100_arabic_questions_train)

arabic_questions_that_dont_have_questionmark = len(df_train[(df_train["lang"] == "ar") & (~df_train["question"].str.startswith("؟"))]["question"].tolist())
print(arabic_questions_that_dont_have_questionmark)
```

```python
### Tokenization of words only consisting of letters, numbers and underscore

arabic_train_split_at_spaces_and_chars_that_are_not_letters = df_train[df_train["lang"] == "ar"]["question"].apply(lambda x: len(re.split(r'\W+', x))).sum()
korean_train_split_at_spaces_and_chars_that_are_not_letters = df_train[df_train["lang"] == "ko"]["question"].apply(lambda x: len(re.split(r'\W+', x))).sum()
telugu_train_split_at_spaces_and_chars_that_are_not_letters = df_train[df_train["lang"] == "te"]["question"].apply(lambda x: len(re.split(r'\W+', x))).sum()

print('Arabic split with seprarate punctuation (train): ', arabic_train_split_at_spaces_and_chars_that_are_not_letters)
print('Korean split with seprarate punctuation (train): ', korean_train_split_at_spaces_and_chars_that_are_not_letters)
print('Korean split with seprarate punctuation (train): ', telugu_train_split_at_spaces_and_chars_that_are_not_letters)

arabic_val_split_at_spaces_and_chars_that_are_not_letters = df_val[df_val["lang"] == "ar"]["question"].apply(lambda x: len(re.split(r'\W+', x))).sum()
korean_val_split_at_spaces_and_chars_that_are_not_letters = df_val[df_val["lang"] == "ko"]["question"].apply(lambda x: len(re.split(r'\W+', x))).sum()
telugu_val_split_at_spaces_and_chars_that_are_not_letters = df_val[df_val["lang"] == "te"]["question"].apply(lambda x: len(re.split(r'\W+', x))).sum()

print('Arabic split with seprarate punctuation (val): ', arabic_val_split_at_spaces_and_chars_that_are_not_letters)
print('Korean split with seprarate punctuation (val): ', korean_val_split_at_spaces_and_chars_that_are_not_letters)
print('Korean split with seprarate punctuation (val): ', telugu_val_split_at_spaces_and_chars_that_are_not_letters)


### Tokenization of words that contain hyphen

arabic_words_that_contain_hyphen_train = df_train[df_train["lang"] == "ar"]["question"].apply(lambda x: len(re.findall(r'\w+-\w+', x))).sum()
korean_words_that_contain_hyphen_train = df_train[df_train["lang"] == "ko"]["question"].apply(lambda x: len(re.findall(r'\w+-\w+', x))).sum()
telugu_words_that_contain_hyphen_train = df_train[df_train["lang"] == "te"]["question"].apply(lambda x: len(re.findall(r'\w+-\w+', x))).sum()

arabic_words_that_contain_hyphen_val = df_val[df_val["lang"] == "ar"]["question"].apply(lambda x: len(re.findall(r'\w+-\w+', x))).sum()
korean_words_that_contain_hyphen_val = df_val[df_val["lang"] == "ko"]["question"].apply(lambda x: len(re.findall(r'\w+-\w+', x))).sum()
telugu_words_that_contain_hyphen_val = df_val[df_val["lang"] == "te"]["question"].apply(lambda x: len(re.findall(r'\w+-\w+', x))).sum()

print('Arabic words that contain hyphen (train): ', arabic_words_that_contain_hyphen_train)
print('Korean words that contain hyphen (train): ', korean_words_that_contain_hyphen_train)
print('Telugu words that contain hyphen (train): ', telugu_words_that_contain_hyphen_train)

print('Arabic words that contain hyphen (val): ', arabic_words_that_contain_hyphen_val)
print('Korean words that contain hyphen (val): ', korean_words_that_contain_hyphen_val)
print('Telugu words that contain hyphen (val): ', telugu_words_that_contain_hyphen_val)


### Tokenize everything neither letter nor whitespace - but all other symbols
### doesn't work well for telugu because it's letters are seen as symbols and are therefore all split up

arabic_num_of_punctuation_train = df_train[df_train["lang"] == "ar"]["question"].apply(lambda x: len(re.findall(r'[^\w\s]', x))).sum()
korean_num_of_punctuation_train = df_train[df_train["lang"] == "ko"]["question"].apply(lambda x: len(re.findall(r'[^\w\s]', x))).sum()
telugu_num_of_punctuation_train = df_train[df_train["lang"] == "te"]["question"].apply(lambda x: len(re.findall(r'[^\w\s]', x))).sum()

arabic_num_of_punctuation_val = df_val[df_val["lang"] == "ar"]["question"].apply(lambda x: len(re.findall(r'[^\w\s]', x))).sum()
korean_num_of_punctuation_val = df_val[df_val["lang"] == "ko"]["question"].apply(lambda x: len(re.findall(r'[^\w\s]', x))).sum()
telugu_num_of_punctuation_val = df_val[df_val["lang"] == "te"]["question"].apply(lambda x: len(re.findall(r'[^\w\s]', x))).sum()

print('Arabic number of punctuation (train): ', arabic_num_of_punctuation_train)
print('Korean number of punctuation (train): ', korean_num_of_punctuation_train)
print('Telugu number of punctuation (train): ', telugu_num_of_punctuation_train)

print('Arabic number of punctuation (val): ', arabic_num_of_punctuation_val)
print('Korean number of punctuation (val): ', korean_num_of_punctuation_val)
print('Telugu number of punctuation (val): ', telugu_num_of_punctuation_val)


arabic_num_of_punctuation_train_ratio = arabic_num_of_punctuation_train / arabic_train_split_at_spaces_and_chars_that_are_not_letters
korean_num_of_punctuation_train_ratio = korean_num_of_punctuation_train / korean_train_split_at_spaces_and_chars_that_are_not_letters
telugu_num_of_punctuation_train_ratio = telugu_num_of_punctuation_train / telugu_train_split_at_spaces_and_chars_that_are_not_letters

arabic_num_of_punctuation_val_ratio = arabic_num_of_punctuation_val / arabic_val_split_at_spaces_and_chars_that_are_not_letters
korean_num_of_punctuation_val_ratio = korean_num_of_punctuation_val / korean_val_split_at_spaces_and_chars_that_are_not_letters
telugu_num_of_punctuation_val_ratio = telugu_num_of_punctuation_val / telugu_val_split_at_spaces_and_chars_that_are_not_letters

print('Arabic number of punctuation ratio (train): ', arabic_num_of_punctuation_train_ratio)
print('Korean number of punctuation ratio (train): ', korean_num_of_punctuation_train_ratio)
print('Telugu number of punctuation ratio (train): ', telugu_num_of_punctuation_train_ratio)

print('Arabic number of punctuation ratio (val): ', arabic_num_of_punctuation_val_ratio)
print('Korean number of punctuation ratio (val): ', korean_num_of_punctuation_val_ratio)
print('Telugu number of punctuation ratio (val): ', telugu_num_of_punctuation_val_ratio)

print('Punctuations in telugu val: ', df_val[df_val["lang"] == "te"]["question"].apply(lambda x: re.findall(r'[^\w\s]', x)).tolist()) # -> each letter in telugu is not seen as a letter, PROBLEM!
```

```python
### Tokenization with NLTK

from nltk import word_tokenize

arabic_train_nltk_count = sum(
    len(word_tokenize(text, language='arabic', preserve_line=True))
    for text in df_train.loc[df_train["lang"] == "ar", "question"] if isinstance(text, str)
)

korean_train_nltk_count = sum(
    len(word_tokenize(text, language='korean', preserve_line=True))
    for text in df_train.loc[df_train["lang"] == "ko", "question"] if isinstance(text, str)
)

telugu_train_nltk_count = sum(
    len(word_tokenize(text, language='telugu', preserve_line=True))
    for text in df_train.loc[df_train["lang"] == "te", "question"] if isinstance(text, str)
)


print('Arabic train nltk word count:', arabic_train_nltk_count)
print('Korean train nltk word count:', korean_train_nltk_count)
print('Telugu train nltk word count:', telugu_train_nltk_count)
```
